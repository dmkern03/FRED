```python
%sql
CREATE OR REPLACE TEMP VIEW temp_view_fred_rates AS
SELECT
  r.series_id,
  r.date,
  r.value,
  m.title,
  m.friendly_name,
  m.units,
  'FRED' AS source
FROM
  investments.fred.silver_rates r
JOIN
  investments.fred.silver_metadata m
ON
  r.series_id = m.series_id;

SELECT *
FROM temp_view_fred_rates;
```
